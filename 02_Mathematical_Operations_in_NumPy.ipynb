# 2. Mathematical Opearations with NumPy

NumPy performs mathematical operations on n-dimensional arrays.

### 2.1 Operations on a single array

```python
import numpy as np

# Create Array
arr = np.array([10, -2, 18, 8, 15, 22, 7])
print("arr:", arr)

# Adding a number to an array
arr1 = arr + 3.5
print("arr1 = arr + 3.5:", arr)

# Subracting a number from an array
arr2 = arr - 1
print("arr2 = arr - 1:", arr2)

# Multiplying array by some number
arr3 = 2 * arr
print("arr3 = 2 * arr:", arr3)

# Dividing array by some number
arr4 = arr / 2
print("arr4 = arr / 2:", arr4)


# Array exponentiation
arr5 = arr ** 2
print("arr5 = arr ** 2", arr5)
```

### 2.2 Operations on Two Arrays

The sizes of arrays must be compatible. The arithmetical operations are performed element wise.

```python
import numpy as np

# Setting a random seed
rng = np.random.default_rng(125)

# Create two arrays
print("\nInitial Arrays:")
arr1 = rng.integers(low=1, high=10, size=5)
arr2 = rng.integers(low=1, high=10, size=5)

print("arr1:", arr1)
print("arr2:", arr2)

# Adding arrays
print("\nAddition:")
arr3 = arr1 + arr2
print("arr3 = arr1 + arr2:", arr3)

# Subtracting arrays
print("\nSubtraction:")
arr4 = arr1 - arr2
print("arr4 = arr1 - arr2:", arr4)

# Multiplying arrays
print("\nMultiplication:")
arr5 = arr1 * arr2
print("arr5 = arr1 * arr2", arr5)

# Arrays division
print("\nDivision:")
arr6 = arr1 / arr2
print("arr6 = arr1 / arr2", arr6)

# Array integer division
print("\nInteger Division:")
arr7 = arr1 // arr2
print("arr7 = arr1 // arr2", arr7)
```

### 2.3 Comparing two arrays

```python
import numpy as np

# Create arrays
arr1 = np.array([1, 6, 3, 1, 7])
arr2 = np.array([7, 6, 8, 4, 3])
arr3 = np.array([7, 6, 8, 4, 3])

# Comparing Arrays
print("Comparing arr1 and arr2. Arrays are equal:", np.array_equal(arr1, arr2))
print("Comparing arr2 and arr3. Arrays are equal:", np.array_equal(arr2, arr3))
```

### 2.4 Aggregation Functions on NumPy Arrays: sum, mean, min, max

```python
import numpy as np

# Create array
arr = np.arange(1, 11)
print("arr:", arr, "\n")

# Sum of array elements
s = arr.sum()
print("Sum of elements of arr:", s)

# Mean value of array elements
m = arr.mean()
print("Mean value of elements of arr:", m)

# Min value of array elements
minimum = arr.min()
print("Min. value of elements of arr:", minimum)

# Max value of array elements
maximum = arr.max()
print("Max. value of elements of arr:", maximum)
```

### 2.5 Trigonometric Functions

### 2.6 Broadcasting in NumPy

The true power of NumPy lies in its fast mathematical operations. The approach that's used by NumPy is to avoid stepping into the Python interpreter by performing element-wise calculations using optimized C code. **Broadcasting** is a clever set of rules that enables fast array calculations for arrays of similar (but not equal!) shapes.

The term **broadcasting** describes how NumPy treats arrays with different shapes during arithmetic operations.

#### Broadcasting Rules

1) When operation on two arrays, NumPy compares their shapes element-wise, it starts with the rightmost dimension and works its way left.
2) Two dimensions are compatible when:
   - they are equal;
   - one of them is one.
3) If the conditions above are not met, a *ValueError: operands could not be broadcast together* exception is thrown. 

```python
# Example 1: Same shape arrays

import numpy as np

arr1 = np.array([
    [1, 2],
    [0, 1]
])

arr2 = np.array([
    [10, 2],
    [8, 1]
])

arr3 = arr1 + arr2
print("\narr1 + arr2 :\n", arr3)

arr4 = arr1 * arr2
print("\narr1 * arr2 :\n", arr4)
```

```python
# Example 2: (3 x 2) and (1 x 2) arrays

import numpy as np

arr1 = np.array([
    [1, 2],
    [2, 0],
    [3, 2]
])

arr2 = np.array([1, 2])

arr3 = arr1 + arr2
print("\narr1 + arr2:\n", arr3)

arr4 = arr1 * arr2
print("\narr1 * arr2:\n", arr4)
```

### 2.7. Project: Partical Simulator

We want to simulate a system containing particles that constantly rotate around a central point at various speeds.

Input: (x0, y0) - starting position of the particle, v - speed of the particle, rotation direction

Wanted: 

1) we want to be able to calculate position of the particles at any instance of time.
2) we want to visualize the movement of the particles

The goal of the project: we want to learn to learn how to use Matplotlib in combination with NumPy

### Solution

In order to solve the first project assignment, we created a *ParticleSimulator* class.
*ParticleSimulator* class accepts Particle objects list at initialization. The method *.evolve()* allows to calculate the position of each particle at any given instance of time.

```python
# Particle Simulator

class Particle:
    def __init__(self, x, y, ang_vel, color="red"):
        self.x = x
        self.y = y
        self.ang_vel = ang_vel
        self.color = color


class ParticleSimulator:
    def __init__(self, particles):
        self.particles = particles

    def evolve(self, dt):
        timestep = 10**(-3)
        nsteps = int(dt/timestep)
        for i in range(nsteps):
            for p in self.particles:
                dx = - p.ang_vel * p.y * timestep
                dy = p.ang_vel * p.x * timestep
                p.x += dx
                p.y += dy

```

### Animation

The code below creates the animation of the Particle Simulator.
Our Particle Simulator contains two particles: 

- p0 (x=0, y=5, ang_vel = pi/30, color="coral");
- p1 (x=3, y=0, ang_vel = pi/180, color="lightgreen");

The evolve time is set to 1 time unit.

The animation function calls the evolve step 90 times.

```python
%matplotlib inline


import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# This line is needed to display the animation in Jupyter Notebook
plt.rcParams["animation.html"] = "jshtml"


# Simulator
p0 = Particle(0, 5, math.pi/30, color="coral")
p1 = Particle(3, 0, math.pi/180, color="lightgreen")
sim = ParticleSimulator([p0, p1])

fig = plt.figure()
ax = plt.subplot()
ax.set_title("Particle Simulator in Action")
ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)

plt.grid()
plt.close()

def init():
    # Zero point
    line = ax.scatter(0, 0, c=["blue"])
    # Initial data
    X = np.array([p.x for p in sim.particles])
    Y = np.array([p.y for p in sim.particles])
    clr = np.array([p.color for p in sim.particles])
    line = ax.scatter(X, Y, c=clr)
    return line, # The comma is important!

def animate(i):
    # We let the particle evolve for 1 timeunit(s)
    sim.evolve(1)
    X = np.array([p.x for p in sim.particles])
    Y = np.array([p.y for p in sim.particles])
    clr = np.array([p.color for p in sim.particles])
    line = ax.scatter(X, Y, c=clr)
    return line,

# Call the animate function each 10 ms
anim = FuncAnimation(fig, animate, 90, init_func=init, blit=True, interval=200)
anim

```
